```python
#Question 2

#M22B23/004
#A98238


import random
import time
import matplotlib.pyplot as plt
import numpy as np

# def swap(a, i, j):
#   a[i], a[j] = a[j], a[i]

# Selection Sort
def selection_sort(arr):
  n = len(arr)
  for i in range(n):
    min_idx = i
    for j in range(i + 1, n):
      if arr[j] < arr[min_idx]:
        min_idx = j
    arr[i], arr[min_idx] = arr[min_idx], arr[i]
  return arr


# # Merge Sort
def merge_sort(arr):
  if len(arr) <= 1:
    return arr
  mid = len(arr) // 2
  left = merge_sort(arr[:mid])
  right = merge_sort(arr[mid:])
  return merge(left, right)

def merge(left, right):
  result = []
  i = j = 0
  while i < len(left) and j < len(right):
    if left[i] < right[j]:
      result.append(left[i])
      i += 1
    else:
      result.append(right[j])
      j += 1
  result.extend(left[i:])
  result.extend(right[j:])
  return result

# Generate random data for sorting to compare the different algorithms- when not given an array
array_sizes = np.arange(100, 1000, 100)
execution_times_selection_sort = []
execution_times_merge_sort = []

for array_size in array_sizes:
  random_array = np.random.randint(1, 100, array_size)

  # Measure execution time for each sorting algorithm
  start_time = time.time()
  selection_sort(random_array.copy())
  selection_time = time.time() - start_time

  start_time = time.time()
  merge_sort(random_array.copy())
  merge_time = time.time() - start_time

  # Store execution times
  execution_times_selection_sort.append(selection_time)
  execution_times_merge_sort.append(merge_time)

# Plotting the execution time
plt.plot(array_sizes, execution_times_selection_sort, color='orange', label='Selection Sort')
plt.plot(array_sizes, execution_times_merge_sort, color='blue', label='Merge Sort')
plt.xlabel('Array Size (n)')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time of two sorting Algorithms')
plt.legend()
plt.show()

"""The line graph shows the execution time of two different sorting algorithms (selection sort and merge sort) as a function of the array size (n). The x-axis of the graph is the array size, and the y-axis is the execution time in seconds. The different colors of the lines represent the different sorting algorithms.
The graph shows that both sorting algorithms have execution times that increase with the array size, but at different rates. Selection sort has the slowest execution time, followed by  merge sort. Merge sort is the fastest of the two algorithms.
The graph also shows that the rate at which the execution time increases with the array size is different for each algorithm. For selection sort, the execution time increases quadratically with the array size (O(n^2)). For merge sort, the execution time increases logarithmically with the array size (O(n log n)).
The graph also shows that the difference in execution time between the two sorting algorithms becomes more pronounced as the array size increases. For example, at an array size of 100, the merge sort algorithm is about twice as fast as the selection sort algorithm. However, at an array size of 1000, the merge sort algorithm is about 10 times as fast as the selection sort algorithm.
Overall, the line graph shows that the merge sort algorithm is the fastest of the two sorting algorithms, and that the selection sort algorithm is the slowest. The graph also shows that the rate at which the execution time increases with the array size is different for each algorithm.
"""

#IMPROVISED LIST
array=[2,5,8,34,11,6,39,7,8,3,4,2,3,7,9,5,1,6,3,4,71,68,43]

sorted_array1=merge_sort(array)
print("Sorted array using mergesort: ",sorted_array1)
print("Execution time for merge sort: ",round(merge_time,3),"seconds")

sorted_array2=selection_sort(array)
print("Sorted array using selection sort: ", sorted_array2)
print("Execution time for selection sort: ",round(selection_time,3),"seconds")


#The most efficient code would be the merge sort because its execution time is less than the selection sort
#The graph also shows how efficient the merge sort algorithm is compared to the selection sort algorithm
#As the input size increases, we see that the time the selection sort takes is longer compared to the merge sort algorithm where the time changes are very insignificant
#This therefore shows that the merge sort is a better algorithm


```
